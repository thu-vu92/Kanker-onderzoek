```python
import numpy as np
import pandas as pd
from scipy.stats import wilcoxon, mannwhitneyu


class Mannwhitney():

    def __init__(self, pvalue = 0.05):
        self.pvalue = pvalue
        self.pvalues = None

    def apply_test(self, pos, neg, column):
        try:
            _, p_value = mannwhitneyu(pos[column].values, neg[column].values, alternative="less")
        except ValueError as e:
            p_value = 0.0
        return p_value

    def fit(self, x, y):
        pos_samples = x[y == 1]
        neg_samples = x[y == 0] 

        self.p_values = np.array(list(map(lambda c: 
            self.apply_test(pos_samples, neg_samples, c), x.columns.values.tolist())))
        return self

    def transform(self, x, y):
        null_rejected = self.p_values < self.pvalue
        # to_delete = [idx for idx, item in enumerate(not_signif) if item == False]
        # return np.delete(x, to_delete, axis = 1), y
        return x.iloc[:,null_rejected], y
```
